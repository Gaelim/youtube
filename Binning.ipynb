```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Desktop/Data/titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
bins = [0,2,5,10,14,17,30,60,100]
labels = ['infant','toddler','young child','older child','teenager','young adult','adult','elderly']
```

```python
df['Age Category'] = pd.cut(df['Age'],bins,labels = labels)
```

```python
df.head(25)
```

```python
df['Age Category'].value_counts().plot(kind='pie')
```

```python
df.to_csv('titanic_category.csv')
```
