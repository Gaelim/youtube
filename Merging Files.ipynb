# Use Merge to Perform Simple Table Joins

```python
import pandas as pda
```

```python
main_table =  pd.read_excel('Desktop\hotel_data.xlsx',sheet_name='2018')
main_table.head()
```

```python
market_table =  pd.read_excel('Desktop\hotel_data.xlsx',sheet_name='market_segment')
meal_table =  pd.read_excel('Desktop\hotel_data.xlsx',sheet_name='meal_cost')
```

```python
merge1 = main_table.merge(meal_table,on="meal")
merge1.head()
```

```python
columns = ['meal','Cost','market_segment','Discount']
#merge1[columns]
```

```python
merge2 = merge1.merge(market_table,on='market_segment')
```
